```python
Iris_df.drop(['Id'],axis=1, inplace=True)
```

```python
sp = {'Iris-setosa':1, 'Iris-virginica':2, 'Iris-versicolor':3}
```

```python
Iris_df.Species = [sp[i] for i in Iris_df.Species]
```

## **Training Modle:**

```python
# Creating X and y variable
X=Iris_df.iloc[:,:4]
y=Iris_df.iloc[:,-1]
```

**4. Decision Tree:**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Initialize the Decision Tree model with max depth
dt_model = DecisionTreeClassifier(max_depth=3)

# Train the model on the training data
dt_model.fit(X_train, y_train)

# Make predictions on the testing data
y_pred_train = dt_model.predict(X_train)

# Evaluate the model on testing data
accuracy_train = accuracy_score(y_train, y_pred_train)
classification_rep = classification_report(y_train, y_pred_train)

print("Accuracy on Training Data : ", accuracy_train)
print("Classification Report:\n", classification_rep)
```

```python
y_pred_test = dt_model.predict(X_test)

accuracy_test = accuracy_score(y_test, y_pred_test)
classification_rep = classification_report(y_test, y_pred_test)

print("Accuracy on Test Data : ", accuracy_test)
print("Classification Report : \n", classification_rep)
```

**Grphical Represetation of Decision Tree Classifier :**

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Initialize the Decision Tree model with max depth
dt_model = DecisionTreeClassifier(max_depth=3)

# Training the model on the training data
dt_model.fit(X_train, y_train)  # Fitting the X & y variable in modle

# Convert class names to strings
class_names = [str(i) for i in Iris_df['Species'].unique()]

# Plot the decision tree
plt.figure(figsize=(12, 8))
plot_tree(dt_model, feature_names=Iris_df.columns[:-1], class_names=class_names, filled=True)
plt.show()
```
